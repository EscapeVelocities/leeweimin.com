```python
#Create dataframe (for total costs) from csv file
nus_costs_df = pd.read_csv('NUS_Costs.csv', index_col='Degree')
```

```python
nus_costs_df.head()
```

```python
#Rank by "Total Cost" column (the lower the cost, the higher the rank)
nus_costs_df['Rank (On Total Cost)'] = nus_costs_df['Total Cost'].rank(method='min')

#Rank by "Number of Years to Graduate" column (the lower the number of years, the higher the rank)
nus_costs_df['Rank (On Number of Years to Graduate)'] = nus_costs_df['Number of Years to Graduate'].rank(method='min')
```

```python
#Dataframe is sorted by total costs with number of years to graduate as tiebreaker
nus_costs_df.sort_values(['Rank (On Total Cost)', 'Rank (On Number of Years to Graduate)'], inplace=True)
```

```python
nus_costs_df.head()
```

```python
#Set size of barplot
plt.figure(figsize=(20,20))
#Create barplot for ranking majors based on total monetary costs
ax = sns.barplot(x='Total Cost', y=nus_costs_df.index, data=nus_costs_df, ci=None)
plt.title('Ranking of Degrees (Based on Total Costs)', loc='left', fontsize='xx-large', fontweight='bold')
plt.show()
```

```python
#Create dataframe (for NUS GES) from xlsx file
nus_ges_df = pd.read_excel('NUS_GES_2017.xlsx', header=[0,1], index_col=0)
```

```python
nus_ges_df.head()
```

```python
#Create Series with gross monthly salaries (25th percentile) of each major
nus_gross_salary_25th = nus_ges_df.loc[:, ('Gross Monthly Salary', '25th Percentile')].copy()
#Series is sorted with highest salaries at the top
nus_gross_salary_25th.sort_values(ascending=False, inplace=True)
```

```python
nus_gross_salary_25th.head()
```

```python
#Set size of barplot
plt.figure(figsize=(20,20))
#Create barplot for ranking majors based on gross monthly salary (25th percentile)
ax = sns.barplot(x=nus_gross_salary_25th, y=nus_gross_salary_25th.index, ci=None)
plt.title('Ranking of Degrees (Based on Gross Monthly Salary - 25th Percentile)', loc='left', fontsize='xx-large', fontweight='bold')
plt.xlabel('Gross Monthly Salary - 25th Percentile')
plt.show()
```

```python
#Get median salaries of each major in a Series
nus_gross_salary_median = nus_ges_df.loc[:, ('Gross Monthly Salary', 'Median')].copy()
#Series is sorted with highest salaries at the top
nus_gross_salary_median.sort_values(ascending=False, inplace=True)
```

```python
nus_gross_salary_median.head()
```

```python
#Set size of barplot
plt.figure(figsize=(20,20))
#Create barplot for ranking majors based on gross monthly salary (median)
ax = sns.barplot(x=nus_gross_salary_median, y=nus_gross_salary_median.index, ci=None)
plt.title('Ranking of Degrees (Based on Gross Monthly Salary - Median)', loc='left', fontsize='xx-large', fontweight='bold')
plt.xlabel('Gross Monthly Salary - Median')
plt.show()
```

```python
#Get 75th percentile salaries of each major in a Series
nus_gross_salary_75th = nus_ges_df.loc[:, ('Gross Monthly Salary', '75th Percentile')].copy()
#Series is sorted with highest salaries at the top
nus_gross_salary_75th.sort_values(ascending=False, inplace=True)
```
