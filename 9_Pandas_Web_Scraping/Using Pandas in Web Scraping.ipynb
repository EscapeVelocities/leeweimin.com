```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page_response = requests.get(f'https://openlibrary.org/authors/OL9388A/William_Shakespeare?page=1')
page_soup = BeautifulSoup(page_response.text, 'html.parser')
```

```python
book_dataset = page_soup.find_all('span', {'class':'resultPublisher'})
book_dataset
```

```python
first_book = book_dataset[0]
number_of_editions = first_book.text.strip()[:3]
number_of_editions
```

```python
first_book = book_dataset[0]
year = first_book.text.strip()[-4:]
year
```

```python
book_dataset = page_soup.find_all('span', {'class':'resultPublisher'})
for book_data in book_dataset:
    number_of_editions = book_data.text.strip()[:3]
    year = book_data.text.strip()[-4:]
    print(number_of_editions)
    print(year)
```

```python
all_books = []
for page_number in range(1, 103):
    page_response = requests.get(f'https://openlibrary.org/authors/OL9388A/William_Shakespeare?page={page_number}')
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    book_titles = page_soup.find_all('h3', {'class':'booktitle'})
    book_dataset = page_soup.find_all('span', {'class':'resultPublisher'})
    for book_title, book_data in zip(book_titles, book_dataset):
        books = {}
        books['book_title'] = book_title.a.string
        books['number_of_editions'] = book_data.text.strip()[:3]
        books['year'] = book_data.text.strip()[-4:]
        all_books.append(books)
print(all_books)
```

```python
all_books_df = pd.DataFrame(all_books)
```
